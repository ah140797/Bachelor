```python
import pandas as pd
import re
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
```

# Vif Analysis on emotion lexicon

```python
df = pd.read_csv('Data/lexicon_analysis/word_diff_emotion_raw_scores.csv')
df
```

```python
X_anx = df.loc[:, ['anticipation','sadness','disgust']]
print(X_anx)
```

```python
X_anx['intercept'] = 1 
vif_anx = pd.DataFrame()
vif_anx['feature'] = X_anx.columns
vif_anx['VIF'] = [variance_inflation_factor(X_anx.values, i) for i in range(X_anx.shape[1])]
print(vif_anx)
```
